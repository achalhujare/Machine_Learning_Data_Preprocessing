```python
#Handling missinh values(using mean and scikit learn)

from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy="mean")
ds[["Age","Salary"]]=si.fit_transform(ds[["Age","Salary"]])
ds
```

```python
#Encoding (oneHotEncoding)

from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
encoded_data=ohe.fit_transform(ds[['Country']])
encoded_data
```

```python
ds['Country'].unique()
```

```python
#Make a new column for array

column_names=ohe.get_feature_names_out(['Country'])
encoded_df=pd.DataFrame(encoded_data,columns=column_names)
encoded_df
```

```python
#join it to original dataset

ds=pd.concat([ds,encoded_df],axis=1)
ds
```

```python
#drop previous columns which are now not necessaryy

ds=ds.drop(columns=['Country'])
ds
```

```python
#Label Encoding

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds["Purchased"]=le.fit_transform(ds["Purchased"])
ds
```

```python
#feature scaling(Reduce large data to smaller)

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
ds[["Age","Salary"]]=sc.fit_transform(ds[["Age","Salary"]])
ds
```

```python
#Divide dataset as Independent (input) and Dependent(output) variables
x=ds[["Age","Salary","Country_France","Country_Germany","Country_Spain"]]
y=ds[["Purchased"]]
```
